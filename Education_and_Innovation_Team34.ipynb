Independent variables with a Pearsons correlation above 0.9 are are said to be collinear. Collinear variables can skew the results of a regression. Based on the correlation matrix above patent_grants and patents_in_force are collinear with one another and then each are collinear with combined_edu_expenditure_total. Because of the low R-squared value and the high p-value of the model with combined_edu_expenditure_total, we will focus only on the models using combined_edu_expenditure_percentage. Once the model is created the Variance Inflation Factor(VIF) will be calculated to see which of the collinear variables (patent_grants and patents_in_force) should be removed.
