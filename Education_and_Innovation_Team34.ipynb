```python
#Updating Location column with actual country names
clean_research_df = research_df
clean_research_df.replace({'LOCATION':iso_country_codes_dict},inplace=True)

#Dropping non total values from SUBJECT column
clean_research_df = clean_research_df[clean_research_df['SUBJECT'] == 'TOT']

#Dropping Headcount measures
clean_research_df = clean_research_df[clean_research_df['MEASURE'] != 'HEADCOUNT']

#Dropping years that are not in the targeted years
clean_research_df = clean_research_df[(clean_research_df['TIME'] > 2015) & (clean_research_df['TIME'] < 2020)]

#Combining INDICATOR and MEASURE columns into new indicator column
clean_research_df['indicator'] = clean_research_df[['SUBJECT', 'INDICATOR', 'MEASURE']].agg('_'.join, axis=1)

#Dropping unneeded columns
clean_research_df = clean_research_df.copy()
clean_research_df = clean_research_df.drop(columns=['INDICATOR','SUBJECT',  'MEASURE', 'FREQUENCY','Flag Codes'])

#Renaming columns
clean_research_df.rename(columns={'LOCATION':'country', 'TIME':'year', 'Value':'value' }, inplace=True)

#Reorder index
clean_research_df = clean_research_df.reindex(columns=['country','year','indicator','value'])

#Resetting the index
clean_research_df.reset_index(drop=True, inplace=True)


clean_research_df.head()
```
